# Data Cleaning

```python
df = pd.read_csv('metadata.csv')
```

```python
df
```

```python
for col in df.columns:
    print(col)
```

```python
cases_df = df[['Municipality code', 'Municipality', 'Cases2007', 'Cases2008', 'Cases2009', 'Cases2010', 'Cases2011', 'Cases2012',
   'Cases2013', 'Cases2014', 'Cases2015', 'Cases2016', 'Cases2017', 'Cases2018', 'Cases2019', 'Population2019']]
```

```python
cases_df = cases_df.sort_values(by = ['Population2019'], ascending = False)
cases_df
```

```python
cases_df['TotalCases'] = cases_df.iloc[:, 2:15].sum(axis = 1)
cases_df = cases_df.sort_values(by = ['TotalCases'], ascending = False)
cases_df
```

```python
cases_df.to_csv('caseData.csv', index=False)
```
